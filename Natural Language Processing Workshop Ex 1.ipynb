```python
my_string = "Let's write RegEx!  Won't that be fun?  I sure think so.  Can you find 4 sentences?  Or perhaps, all 19 words?"
```

#### How do you split a string into sentences?
- notice where is the punctuation

```python
sentence_ending = r"[.?!]"

# Split my_string on sentence endings and print the result
print(re.split(sentence_ending, my_string))
```

```python
# Find all capitalized words in my_string and print the result
capitalized_words = r"[A-Z]\w+"
print(re.findall(capitalized_words, my_string))
```

```python
# Split my_string on spaces and print the result
spaces = r"[\s+]"
print(re.split(spaces, my_string))

token = r"\w+"
print(re.findall(token, my_string))
```

```python
# Find all digits in my_string and print the result
digit = r"[\d]"
print(re.findall(digit, my_string))
digits = r"\d+"
print(re.findall(digits, my_string))
```

```python
my_string = re.sub(digit, " digit ", my_string)
print(my_string)
```

## Word Tokenizing with NLTK

```python
############### LOADING TEXT ###############
shakespeare = ""

with open("shakespeare.txt",'r') as file:
    file.readline()
    for line in file:
        data = str(line).strip()
        shakespeare = shakespeare + data + ' '
        
shakespeare = shakespeare.lower()
```

## 1. Tokenization

â€“ whitespace, newline, blankline, word, treebank, sexpr, regexp,
Punkt sentence segmenter

```python
############### Word Tokenize vs Sent Tokenize ###############

# Import necessary modules
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize

# How many words in this text file?
ss_wordtokens = word_tokenize(shakespeare)
# print(ss_wordtokens)
# print(len(ss_wordtokens))

# How many senetences in this text file?
ss_senttokens = sent_tokenize(shakespeare)
# print(ss_senttokens)
# print(len(ss_senttokens))

# How many Unique Words are there?
unique_tokens = set(word_tokenize(shakespeare))
print(len(unique_tokens))
```

```python
############### Regexp Tokenizer ###############
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')

ss_regword_tokens = tokenizer.tokenize(shakespeare)

# print(ss_regword_tokens)
# print(len(ss_regword_tokens))

# What is the difference with WordTokenizer and RegexpTokenizer?
```

## 2. Stopwords + Punctuation

```python
############### STOPWORDS ###############
# What are stopwords? 

# Import necessary modules
from nltk.corpus import stopwords

stop_words = stopwords.words('english')

print(type(stop_words))
print(stop_words)

# Are there anymore stopwords not included?
# Are you able to add or remove your own stopwords?

stop_words.append('thy')
print(stop_words[-1])

stop_words.remove('i')
print(stop_words)
```

```python
# Now you have to remove the stopwords from the content
# After removing, how many words are there left?
clean_text = []

print(len(ss_regword_tokens))

for w in ss_regword_tokens:
    if w not in stop_words:
        clean_text.append(w)
        
print(len(clean_text))
```

## 3. Stemming vs Lemmatization

â€“ Porter, Lancaster, regexp, Snowball


```python
############### Stemming ###############
# What is stemming?

from nltk.stem import PorterStemmer
from nltk.stem.lancaster import LancasterStemmer

porter_stemmer = PorterStemmer()

#how does stemmer words?
test_words= ["cement", "owed", "maximum", "community"]
# test_words= ["understand", "understands", "understanding"]

for w in test_words: 
    print(porter_stemmer.stem(w))

# Other Forms for Stemmer
lancaster_stemmer = LancasterStemmer()
for w in test_words: 
    print(lancaster_stemmer.stem(w))

    
```

```python
############### LEMMATIZATION ###############
# What is the difference with Stemming and Lemmatization

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

test_words= ["churches", "abaci","community", "are", "is"]

for w in test_words: 
    print(lemmatizer.lemmatize(w))
    
# WordNet superficially resembles a thesaurus, but...

print(lemmatizer.lemmatize('is', pos='v'))
print(lemmatizer.lemmatize('are', pos='v'))
print(lemmatizer.lemmatize('better', pos='a'))
print(lemmatizer.lemmatize('run'))
```

## 4. Frequency Distribution

```python
############### Visualizing with NLTK ###############
from nltk import FreqDist

fd = FreqDist(clean_text)
fd.plot()
fd.plot(50, cumulative=True)
fd.most_common(12)
```

```python
############### Visualizing with WORDCLOUD ###############
import matplotlib.pyplot as plt
from wordcloud import WordCloud,STOPWORDS

def wordcloud_draw(data, color = 'black'):
    words = ' '.join(data)
    wordcloud = WordCloud(stopwords=STOPWORDS,
                      background_color=color,
                      width=2500,
                      height=2000
                     ).generate(words)
    plt.figure(1,figsize=(13, 13))
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
    #plt.savefig('WordCloud.png', dpi=300)
    
wordcloud_draw(clean_text,'white')
```
